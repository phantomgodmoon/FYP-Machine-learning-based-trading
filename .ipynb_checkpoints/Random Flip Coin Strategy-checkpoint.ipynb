```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
SPY = yf.download("SPY",start="2017-01-01", end="2022-08-29")
```

```python
SPY.pop("Volume")
SPY.pop("High")
SPY.pop("Low")
SPY.pop("Adj Close")
```

```python
change_arr=[]
for i in range(1423):
    change_arr.append(SPY.iat[i,1]/SPY.iat[i,0]-1)
    
SPY['Change (%)']=change_arr
```

```python
print(SPY)
length=len(SPY)
print(length)
```

```python
i=np.random.choice([0, 1], size=(length), p=[0.5, 0.5])
df = pd.DataFrame(data=i, dtype=np.int8)
df= df.T
```

```python
for i in range(9999):
    arr=np.random.choice([0, 1], size=(length), p=[0.5, 0.5])
    df=df.append(pd.DataFrame(arr).T)
    
```

```python
df
```

```python
df=df.append(pd.DataFrame(change_arr).T)

df_2=df.copy()
df_3=df.copy()
df_4=df.copy()

for i in range(10000):
    change=df.iat[10000,0]
    if(df.iat[i,0]==0):
        df.iat[i,0]=1-change
    else:
        df.iat[i,0]=1+change

        
        
```

```python
for j in range (1,length):
    change=df.iat[10000,j]
    for i in range(10000):
        if(df.iat[i,j-1]<0):
            df.iat[i,j]=0
        else:
            if(df.iat[i,j]==0):
                df.iat[i,j]=df.iat[i,j-1]*(1-change)
            else:
                df.iat[i,j]=df.iat[i,j-1]*(1+change)

df
```

```python
return_arr= df[length-1]
return_arr
```

```python
return_arr=np.sort(return_arr)
```

```python
plt.title("Expected Return of the Random Trading Strategy")
plt.boxplot(return_arr,vert=False)

plt.show()
```

<h1> ii) Trading Strategy After taking commision fee and spread price in account.</h1> <br>
Take commission fee and spread price gap = <b>0.001
    

```python
commission_fee=0.001
commission_adjust=1-commission_fee
```

```python
for i in range(10000):
    change=df_2.iat[10000,0]
    if(df_2.iat[i,0]==0):
        df_2.iat[i,0]=(1-change)*commission_adjust
    else:
        df_2.iat[i,0]=(1+change)*commission_adjust
        

               
```

```python
for j in range (1,length):
    change=df_2.iat[10000,j]
    for i in range(10000):
        if(df_2.iat[i,j-1]<0):
            df_2.iat[i,j]=0
        else:
            if(df_2.iat[i,j]==0):
                df_2.iat[i,j]=df_2.iat[i,j-1]*(1-change)*commission_adjust
            else:
                df_2.iat[i,j]=df_2.iat[i,j-1]*(1+change)*commission_adjust
                
```

```python
df_2
```

```python
return_arr_2= df_2[length-1]
return_arr_2
return_arr_2=np.sort(return_arr_2)
plt.title("Expected Return of the Random Trading Strategy, \n taking commission fee into account")
plt.boxplot(return_arr_2,vert=False)

plt.show()
```
