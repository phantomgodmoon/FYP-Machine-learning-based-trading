```python
from arch import arch_model
from arch.unitroot import *
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
def ADF_test(target_list):
    num_ticker=len(target_list)
    completed_list=np.array([["",np.float64(0),np.float64(0)]])
    for target in target_list:
        data=yf.download(target,start='2018-01-01',end='2022-11-01')
        model=ADF(data['Adj Close'], trend = 'ct')
        print(model.summary())
        model_result=np.array([[target,model.stat,model.pvalue]])
        completed_list=np.concatenate((completed_list,model_result))
    completed_list=completed_list[1:,:]
    return completed_list
```

```python
def PHP_test(target_list):
    num_ticker=len(target_list)
    completed_list=np.array([["",np.float64(0),np.float64(0)]])
    for target in target_list:
        data=yf.download(target,start='2018-01-01',end='2022-11-01')
        model=PhillipsPerron(data['Adj Close'], trend = 'ct')
        print(model.summary())
        model_result=np.array([[target,model.stat,model.pvalue]])
        completed_list=np.concatenate((completed_list,model_result))
    completed_list=completed_list[1:,:]
    return completed_list
```

```python
def KPSS_test(target_list):
    num_ticker=len(target_list)
    completed_list=np.array([["",np.float64(0),np.float64(0)]])
    for target in target_list:
        data=yf.download(target,start='2018-01-01',end='2022-11-01')
        model=KPSS(data['Adj Close'], trend = 'ct')
        print(model.summary())
        model_result=np.array([[target,model.stat,model.pvalue]])
        completed_list=np.concatenate((completed_list,model_result))
    completed_list=completed_list[1:,:]
    return completed_list
```

```python
adf=ADF_test(["SPY","^HSI","^FTSE","000001.SS","^GDAXI","^N225","BTC-USD","DX-Y.NYB","CL=F","NG=F","GC=F"])
php=PHP_test(["SPY","^HSI","^FTSE","000001.SS","^GDAXI","^N225","BTC-USD","DX-Y.NYB","CL=F","NG=F","GC=F"])
kpss=KPSS_test(["SPY","^HSI","^FTSE","000001.SS","^GDAXI","^N225","BTC-USD","DX-Y.NYB","CL=F","NG=F","GC=F"])
```

```python
df_adf = pd.DataFrame(adf, columns = ['Ticker','Test Statistics','P-value'])
df_php = pd.DataFrame(php, columns = ['Ticker','Test Statistics','P-value'])
df_kpss = pd.DataFrame(kpss, columns = ['Ticker','Test Statistics','P-value'])
```
